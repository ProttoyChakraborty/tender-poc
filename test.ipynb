```python
from utils.sotr_construction import SOTRMarkdown
from utils.llm_client import  LLMClient
from dotenv import  load_dotenv
import os 
```

```python
load_dotenv()
```

```python
project_id=os.getenv("PROJECT_ID")
location=os.getenv("LOCATION")
model=os.getenv("MODEL")
```

```python
file_path="./docs/SOTR-Cleaned-V4.pdf"
file_id="test1"
```

```python
llm_client=LLMClient(project_id=project_id,location=location,model=model)
```

```python
sotr=SOTRMarkdown(llm_client=llm_client)
```

```python
sotr.load_from_pdf(file_path,file_id)
```

```python
sotr.markdown_text
```

```python
split_text=sotr.get_matrix_points()
```

```python
split_text
```

```python
cleaned_csv_data=["Sr. No.","Clause","Clause Reference"]
for i,row in enumerate(split_text[1:]):
    items=row.split("|")
    cleaned_csv_data.append([i,items[1].replace('"',''),items[2]])
```

```python
cleaned_csv_data
```

```python
import csv
```

```python
with open('compliance_matrix.csv', 'w') as f:
    
    # Create a CSV writer object that will write to the file 'f'
    csv_writer = csv.writer(f)
    # Write all of the rows of data to the CSV file
    csv_writer.writerows(cleaned_csv_data)
```
